```python
#df = df.dropna() Removing Null value data rows
imputer = SimpleImputer(missing_values=np.nan,strategy='most_frequent')
imputer.fit(df.iloc[:,0:13])
df.iloc[:,0:13] = imputer.transform(df.iloc[:,0:13])

# it is same as doing

# df["Column_Name"].fillna(df["Column_Name"].value_counts().index[0],inplace=True) 

# As value_counts give the descending array of value's occurrence where first is most frequently occuring.

# Or we can make our own category also in place of null values

# df["Column_Name"].fillna("Empty") 

#So it is not null but a new category

#One more way to handle null values is interpolate. This will fill the value with according to above and down row of missing value

# df["Column_Name"].interpolate(inplace=True)
```
